Firstly, we had to acquire our data. To do this, we looked at Roboflow Universe and found a labeled dataset that had pistols, which we downloaded and uploaded to Roboflow.



Luckily for us, the dataset was already labeled, so we did not need to assign parts of the data to each of us for labeling. This would otherwise be standard procedure for using Roboflow with an unlabeled dataset. Afterwards, you would look at your assigned images and label items in the images (in our case, pistols).

We now need to generate our dataset and ensure that our data has the correct formats. Firstly, we select to separate our data into a training set, validation, and test set. The validation set is essentially a test set that is reused after each epoch to give us an idea of how effective our model currently is and to enable early stopping if we are over or underfitting.




Thereafter, we standardize our images to ensure they have the same amount of pixels and are oriented correctly.




Finally, we can augment our dataset to challenge our model further by disturbing something such as the orientation, the saturation, or the noise level. This can ensure that our model becomes better at detecting the object even if the data has a tendency to, for example, have some dead pixels or have different levels of saturation because they were taken with different cameras.



In our case, we just wanted to exemplify how to use Roboflow and therefore did not add any noise.


Then a model is created using the dataset, which you can either download and run on your own computer or let Roboflow run it. We tried running it on our own PC, which can be seen in the picture below. However, as can be seen on the image GPU_mem is 0 which means I don't have a graphics card. Therefore it took far longer and produced worse results.



For this reason, we let Roboflow do the heavy lifting and after less than 2 hours, it had produced a model with 87.8% precision.



If we take a look at the related graphs, we can see the mean average precision steadily rises until it flattens out at around 90%. The object loss and class loss aren't really relevant with only one class, but the Box Loss follows a similar-looking curve. This flattening of the curve indicates that the model is not underfitted.



Furthermore, by comparing the graph from the training with the AP for the test images, we can see that it actually has a higher average precision, which also indicates that the model has not been overfitted.



You could argue that an average precision of around 90% is not extremely high, but we evaluated it to be enough for this example project.
